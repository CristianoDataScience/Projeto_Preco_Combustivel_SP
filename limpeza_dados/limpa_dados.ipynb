## Importando base de dados

```python
df = pd.read_csv('https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/shpc/dsan/2022/precos-gasolina-etanol-04.csv', sep=';')
df.head(3)
```

```python
# Limpeza dos dados e colunas da base
df.drop(['Regiao - Sigla', 'Revenda', 'CNPJ da Revenda', 'Nome da Rua', 'Numero Rua', 'Complemento', 'Cep', 'Valor de Compra', 'Bandeira'], axis=1, inplace=True)
```

```python
df.drop_duplicates().dropna()
```

```python
# Renomendo a coluna "Estado-Sigla" para "Estado"
df.rename(columns={'Estado - Sigla': 'Estado'}, inplace=True)
```

```python
df
```

```python
# Obtendo informações do DataFrame
df.dtypes
```

```python
df['Valor de Venda'] = df['Valor de Venda'].str.replace(',','.').astype(float)
```

```python
df
```

```python
# Tranformando data no formato correto 
df['Data da Coleta'] = pd.to_datetime(df['Data da Coleta'])
```

```python
# Salvando base limpa formato "CSV"
df.to_csv('dados_limpos.csv', index=False)
```
